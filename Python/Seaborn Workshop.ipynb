```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
import seaborn as sns
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.set_theme(style="darkgrid")
tips.to_csv("tips_dataset.csv",index=False)
import pandas as pd
import os
os.getcwd()
{"type":"string"}
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))

sns.scatterplot(data=tips,x="total_bill",y="tip",hue="time",size="size",palette="deep")
plt.title("scatterplot of total bill vs tip")
plt.show()
```

```python
sns.lineplot(data=tips,x='size',y='total_bill',hue='sex',markers='o')
plt.title("lineplot of the totalbill vs size")
plt.show()
```

# 3. Bar plot

```python
sns.barplot(data=tips,x='day',y='total_bill',hue='sex',palette='muted'
)
plt.title("Barplot of totalbill byday")
plt.show()
```

# 4. Box plot

```python
sns.boxplot(data=tips,x='day',y='tip',hue='smoker',palette='Set2')
plt.title("day tips")
plt.show()
```

# 5.violen plot

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='time',split=True,
palette='pastel')
plt.title("violen plot of total bill by day and time")
plt.show()
```

# 6. Column plot

```python
sns.countplot(data=tips,x='day',hue='smoker',palette='dark')
title=("count plot of days by smoker status")
plt.show()
```

# 7. Regression plot

```python
sns.regplot(data=tips,x='total_bill',y='tip',scatter_kws={'s':50,'color':'red'})
plt.title("regression plot of total bill vs tips")
plt.show()
```

# 8. Histogram plot

```python
sns.histplot(data=tips,x='total_bill',bins=20,kde=True,color='blue')
plt.title("histogram of total bill with KDE")
plt.show()
```

# 9. Pair plot

```python
sns.pairplot(tips,hue='sex',vars=["total_bill","tip","size"],palette='husl')
plt.title("pair plot : numeric variables by gender",y=1.02)
plt.show()
```

# 10. Catplot

```python
sns.catplot(data=tips,x='day',y='tip',hue='sex',kind='point',palette='bright')
plt.title("catplot(point):tips by day and gender")
plt.show()
```

# 11. Jointplot

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter',
hue='smoker', color='purple', palette='coolwarm')
plt.suptitle("Jointplot: Total Bill vs Tip", y=1.02)
plt.show()
```

# 12.Facegrid

```python
g = sns.FacetGrid(tips, col='time', row='smoker',
margin_titles=True).map(sns.histplot, 'total_bill', bins=20,
kde=True).add_legend()
g
```

# 13. strip plot

```python
sns.stripplot(data=tips, x='day', y='tip', hue='sex', jitter=True,
palette='Set1')
plt.title("strip plot: Tips by data and gender")
plt.show()
```

# 14. KDE plot

```python
sns.kdeplot(data=tips, x='total_bill',hue='sex', fill=True,
palette='tab10')
plt.title("kde plot:Total bill density by gender")
plt.show()
```
